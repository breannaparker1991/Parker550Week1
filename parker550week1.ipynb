```python
import pandas as pd
import numpy as np
```

```python
#dataframe from csv
df = pd.read_csv("video_games.csv")
```

```python
#show first 10 lines of data
df.head(11)
```

```python
#number of rows and columns
df.shape
#looks like there are 16 columns and 16719 rows. This means that there is quite a lot of data. 
```

```python
#find stop 5 games by critic score
df.nlargest(5, ['Critic_Score'])
```

```python
#Find the first five games in the data frame on the SNES platform
display(df[df['Platform']=='SNES'].iloc[:5])
```

```python
#sum of each group of publishes
global_sales = df.groupby(by=['Publisher'])['Global_Sales'].sum()
#sort values by highest to lowest
global_sales.sort_values(ascending=False, inplace = True)
#display top 5
display(global_sales.iloc[:5])
```

```python
#Create a new column in the data frame that calculates the percentage of global sales from North America. 
df['percentage_sales'] = df['NA_Sales'] / df['Global_Sales'] * 100 
#Display the first five rows of the new data frame.
display(df.iloc[:5])
```

```python
#Find the number NaN entries (missing data values) in each column.
display(df.isnull().sum())
```

```python
#checking types of the fields
df.dtypes
```

```python
#convert all values to numeric
df['User_Score'] = pd.to_numeric(df['User_Score'])
#print median value
print(df['User_Score'].median())
```

```python
#replace all blanks with median score
df['User_Score'] = df['User_Score'].fillna(df['User_Score'].median())
#print new values
display(df['User_Score'])
```
